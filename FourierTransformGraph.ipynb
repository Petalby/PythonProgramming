```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Parameters for the Fourier series
T = 2*np.pi            # Period of the square wave
n_terms = 20             # Number of Fourier series terms
t = np.linspace(-T, T, 1000)  # Time vector

# Fourier series approximation of a  wave

a_o = np.pi/2
def fourier_series_summation(t, T, n_terms):
    result = np.zeros_like(t)
    for n in range(1, n_terms):  
        term = -(1 / ( n)) * np.sin(n*t)
        result += term
    return result

#Fourier series approximation
wave_approx = a_o + fourier_series_summation(t, T, n_terms)

# Plot the Fourier series approximation
plt.figure(figsize=(10, 5))
plt.plot(t, wave_approx, label=f'{n_terms} terms')
plt.title(f'Fourier Series Approximation of a Wave with {n_terms} Terms')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```
